#### Web Scrapping a Wikipedia Article

```python
def make_soup(url):
    html = requests.get(url).content
    return BeautifulSoup(html, 'lxml')
#soup = BeautifulSoup(html, 'lxml')
#return soup
```

```python
soup = make_soup('https://en.wikipedia.org/wiki/List_of_police_dog_breeds')
soup.prettify()
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_police_dog_breeds'
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
soup.prettify()
```

```python
soup.find_all('li')
```

```python
[i.text for i in soup.find_all('li') if re.match('[A-Z]', i.text)]
```

```python
dogs = [i.text for i in soup.find_all('li') if re.match('[A-Z]', i.text)]
dogs[0:17]
```
