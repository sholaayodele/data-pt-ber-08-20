## Investigating and Replacing null values



```python
# df.null() checks the colum and returns which is empty(Nan) with True/False.
## df.isnull()

# Sum the columns: False = 0 ( No missing value ), True = 1(There is missing value)
## df.isnull().sum()

# Check which column has no missing value (0)
df.isnull().sum() == 0

# Extract the columns which has missing values using WHERE function 
## null_cols = df.isnull


#pd.isnull(df['charge120'])

#df.isnull().sum() / len(df)
```

```python
# Columns with missing values
null_cols = df.isnull().sum()
#null_cols[null_cols > 0]

## Give me null_cols with no missing values
## null_cols = df.isnull().sum()
## null_cols[null_cols == 0]

### Give me null_cols where it is greater 0
### null_cols = df.isnull().sum()
### null_cols[null_cols > 0]

## Columns with missing values more than 10000
null_cols[null_cols > 10000]
```

```python
# count of none missing values of all coulumns

df.count()
```

```python
# Drop column evMotor(Has a lot of missing values).. General rule for dropping columns 
df.drop('evMotor', axis = 1)

## One can pass a list of all columns to be dropped
##df.drop(['evMotor', 'mfrCode'], axis = 1)
```

```python
# List of all colums with more than 10000 missing values

cols_missing_10000 = list(null_cols[null_cols > 10000].index)
print(cols_missing_10000)
```

```python
## Drop the column with more than 100000 missing values
df.drop(cols_missing_10000, axis = 1)
```

```python
# Another way of dealing with missing Values
## Fuel type: Electricity! This means that they will have no cylinders and no displacement. replcace by 0!

sum(df['cylinders'].isnull())
```

```python
# Fill the missing values with 0

df['cylinders'] = df['cylinders'].fillna(0)
```

```python
sum(df['cylinders'].isnull())
```

## Cleaning Text

```python
# Find the distinct values for the column trany. Thats to put the column into a set.
set(df['trany'])
```

```python
## Clean the trany column. First replace the "-", ""
## df['trany'].str.replace("-", "")
df['trany'] = df['trany'].str.replace("-", "")

# Replace Automatic with Auto
#df['trany'] = df['trany'].str.replace("Automatic", "Auto")
```

```python
df['trany'] = df['trany'].str.replace("Automatic", "Auto").str.replace(" ", "")
set(df['trany'])
```
