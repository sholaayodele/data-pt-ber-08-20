### DataFrame Aggregation using group by



```python
df.mean()
```

```python
##group by Fuel Type

df.groupby(['Fuel Type']).mean()
```

```python
df.groupby(['Fuel Type'])['Cylinders'].mean()
```

```python
df.groupby(['Fuel Type'])['Cylinders', 'City MPG'].mean().reset_index()
```

```python
## df.groupby(['Fuel Type'])['Cylinders'].mean()
```

```python
df.groupby(['Fuel Type'])['Cylinders', 'City MPG'].agg(['mean', 'max', 'min'])
```

```python
##resetting the abovr to a data frame 

df.groupby(['Fuel Type'])['Cylinders', 'City MPG'].agg(['mean', 'max', 'min']).reset_index()
```
