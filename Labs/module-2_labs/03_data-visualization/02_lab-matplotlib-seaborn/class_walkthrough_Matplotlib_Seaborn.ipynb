## Matplotlib Refresher

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette('husl')  # TO MAKE THINGS PRETTY
```

```python
help(sns.set_palette)
```

```python
df = pd.DataFrame()
df['x'] = np.arange(0,10,1)
df['y'] = df['x']**2
print(df)
```

```python
plt.plot(df['x'], df['y'])
```

```python
sns.lineplot(df['x'], df['y'])
```

```python
help(sns.lmplot)
```

```python
sns.lmplot('x', 'y', data = df)
```

```python
sns.lmplot('x', 'y', data = df, fit_reg = True)
```

## Seaborn

##### USING REAL DATA  

```python
data = pd.read_csv('vehicles.csv')
data
```

### Box Plot

#### In Matplotlib

```python
data.boxplot(column = 'Combined MPG')
```

```python
print(data['Combined MPG'].mean())
print(data['Combined MPG'].median())
```

### In Seaborn

```python
sns.boxplot('Combined MPG', data = data)

# sns.boxplot(x = 'Combined MPG', y = 'Highway MP' data = data)  This to compare two columns
```

#### Violinplot

```python
sns.violinplot('Combined MPG', data = data)
## sns.violinplot(x = 'Combined MPG', y = 'Highway MPG' data = data)   ## Plotting two columns together
```

#### Histogram

##### In Matplotlib

```python
plt.hist(data['Combined MPG'])
```

```python
sns.distplot(data['Combined MPG'], bins = 50)
```

##### Bar plot

```python
help(sns.barplot)
```

```python
fig = sns.barplot(x = 'Highway MPG', y = 'Fuel Type', data = data, estimator = np.median, saturation =0.9).get_figure()
fig.savefig('example.png')
```
